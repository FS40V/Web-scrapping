```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://www.passiton.com/inspirational-quotes"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
data = BeautifulSoup(page, 'html.parser')
print(data)
```

```python
# find all the sections with specifiedd class name
cards_data = data.find_all('div', attrs={'class', 'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'})


# total number of cards
print('Total Number of Cards Found : ', len(cards_data))

# source code of hotel cards
#for card in cards_data:
#    print(card)

for card in cards_data:
    category = card.find('h5', attrs={'class', 'value_on_red'})
    link = card.find('h5', attrs={'class', 'value_on_red'})
    imge = card.find('a')
    print(imge.img['alt'],imge.img["src"],category.a.text,link.a.get('href'))
    
```

```python
import pandas as pd

# create a list to store the data
scraped_data = []

for card in cards_data:

    # initialize the dictionary
    card_details = {}

    # get the hotel name
    category = card.find('h5', attrs={'class', 'value_on_red'})
    category = category.a.text
    link = card.find('h5', attrs={'class', 'value_on_red'})
    link = link.a.get('href')
    imge = card.find('a')
    imge = imge.img["src"]
    lines = card.find('a')
    lines = lines.img['alt']

    # add data to the dictionary
    card_details['theme'] = category
    card_details['url'] = link
    card_details['image'] = imge
    card_details['lines'] = lines
    
    
    # append the scraped data to the list
    scraped_data.append(card_details)

# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)

print(dataFrame)

# save the scraped data as CSV file
dataFrame.to_csv('D:\\Farkhanda_DS_Task1\\q1.csv', index=False)
```
